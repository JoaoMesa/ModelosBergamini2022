```python
from docplex.mp.model import Model

m1 = Model(name='Modelo pg 214')

itens = 3
periodos = 2
microperiodos = 4

x = {(i, m): m1.integer_var(name='x{0}{1}'.format(i+1, m+1)) for i in range(itens) for m in range(microperiodos)}
y = {(i, m): m1.binary_var(name='y{0}{1}'.format(i+1, m+1)) for i in range(itens) for m in range(microperiodos)}
z = {(i, m): m1.binary_var(name='z{0}{1}'.format(i+1, m+1)) for i in range(itens) for m in range(microperiodos)}
I = {(i, t): m1.integer_var(name='I{0}{1}'.format(i+1, t+1)) for i in range(itens) for t in range(periodos)}

C = [280, 400]
s = [350, 100, 90]
d = [[1, 10], [2, 4], [2, 8]]
h = [150, 100, 70]
b = [20, 10, 20]


#DLS 7 - Função objetivo
m1.minimize( m1.sum(s[i] * y[i,m] for i in range (itens) for m in range(microperiodos)) + 
            m1.sum (h[i] * I[i,t] for i in range(itens) for t in range(periodos)))

#DLS 8 

for i in range(itens):
    for t in range(periodos):
        if(t == 0):
            m1.add_constraint(m1.sum(x[i,m] for m in range(microperiodos)) - d[i][t] == I[i,t]);
        else:
            m1.add_constraint(I[i, t-1] + m1.sum(x[i,m] for m in range(microperiodos)) - d[i][t] == I[i,t])

# DLS 9

for i in range(itens):
    for t in range(periodos):
        for m in range(microperiodos):
            m1.add_constraint(b[i] * x[i,m] <= C[t] * z[i,m])

#DLS 10

for t in range(periodos):
    m1.add_constraint(m1.sum(b[i]*x[i,m] for i in range(itens) for m in range(microperiodos)) <= C[t])

#DLS 11

for m in range(microperiodos):
    m1.add_constraint(m1.sum(z[i,m] for i in range(itens)) <= 1)

#DLS 12

for i in range(itens):
    for m in range(microperiodos):
        if m != 0:
            m1.add_constraint(y[i,m] >= z[i,m] - z[i,m-1])
    
        

print("OK\n")
```

```python
print(m1.export_as_lp_string())
```

```python
sol = m1.solve()
print(sol)

```
